# Asset Class Trend Following 策略回測 (equityV2)\n\n本實作紀錄從 **2026-01-02** 開始的績效，並在 **2026-04-01** 起自動切換選股池，納入新增標的。

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from backtest_equityV2 import clean_data, BacktesterV2, calculate_metrics

# 1. 讀取並載入已經過前處理的 2026Q2 數據檔
# 該檔案包含價格與成交量，且已完成缺失值補齊。
prices, volumes, code_to_name = clean_data('資料26Q2-1.xlsx')

# 2. 設定策略參數 (依據 equityV 標準)
# SMA 303, ROC 14, 停損比例 9.99%, 再平衡週期 9 天。
sma, roc, sl, reb = 303, 14, 0.0999, 9

# 3. 初始化並執行回測引擎
# BacktesterV2 已針對日期邏輯進行封裝，2026/04/01 起會自動擴充選股範圍。
bt = BacktesterV2(prices, volumes, code_to_name)
eq, trades, hold, trades2, daily = bt.run(sma, roc, sl, reb, 'peak', 10)

# 4. 篩選記錄期間：從 2026/01/02 開始記錄以方便驗證
mask = (eq['日期'] >= '2026-01-02')
res_p = eq[mask]

# 5. 可視化：繪製 2026 權益曲線
plt.figure(figsize=(12, 6))
plt.plot(res_p['日期'], res_p['權益'], color='blue', linewidth=2)
plt.title('Equity Curve (equityV2 - From 2026/01/02)')
plt.xlabel('Date (日期)')
plt.ylabel('Equity (權益淨值)')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# 6. 輸出期間核心指標
cagr, mdd, calmar, total_ret = calculate_metrics(res_p)
print(f"年化報酬率 (CAGR): {cagr:.2%}")
print(f"最大回撤 (MaxDD): {mdd:.2%}")
print(f"Calmar Ratio: {calmar:.2f}")
print(f"總報酬率: {total_ret:.2%}")
```
